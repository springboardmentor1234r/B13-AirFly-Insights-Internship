# Missing Values Analysis

```python
df.isnull().sum()  #  Several columns contain missing values, particularly time-related columns and delay-related columns
```

**Handle Delay Columns**

```python
# These columns should be 0 if no delay:

delay = [
    'DEP_DELAY', 'ARR_DELAY',
    'DELAY_DUE_CARRIER', 'DELAY_DUE_WEATHER',
    'DELAY_DUE_NAS', 'DELAY_DUE_SECURITY',
    'DELAY_DUE_LATE_AIRCRAFT'
]

for col in delay:
    df[col] = df[col].fillna(0)
```

Missing values in delay columns were replaced with 0 because absence of delay indicates zero delay minutes.

**Handle Cancellation Code**

```python
df['CANCELLATION_CODE'] = df['CANCELLATION_CODE'].fillna('Not Cancelled')
```

Flights that were not cancelled were labeled as "Not Cancelled".

**Time Columns**

```python
# Flights that are cancelled will have missing times.

# We DO NOT fill them with 0.
# Instead we keep them as NaN

time= [
    'DEP_TIME', 'ARR_TIME',
    'TAXI_OUT', 'TAXI_IN',
    'WHEELS_OFF', 'WHEELS_ON',
    'ELAPSED_TIME', 'AIR_TIME'
]

df[time] = df[time].copy()
```

**Convert Data Types Properly**

```python
# Convert categorical columns
cat_cols = ['AIRLINE_CODE', 'ORIGIN', 'DEST', 'ORIGIN_CITY', 'DEST_CITY']

for col in cat_cols:
    df[col] = df[col].astype('category')

#  Convert numeric small integers:
df['CANCELLED'] = df['CANCELLED'].astype('int8')
df['DIVERTED'] = df['DIVERTED'].astype('int8')
```

**Feature Engineering**

```python
# Departure Hour
df['DEP_HOUR'] = (df['CRS_DEP_TIME'] // 100).astype('int')
```

```python
# Arrival Hour
df['ARR_HOUR'] = (df['CRS_ARR_TIME'] // 100).astype('int')
```

Departure and arrival hours were extracted to analyze time-based flight patterns

```python
# Create Route
df['ROUTE'] = df['ORIGIN'].astype(str) + '-' + df['DEST'].astype(str)
```

A route column was created to identify flight origin-destination pairs.

```python
# Delay Flag
df['IS_DELAYED'] = np.where(df['ARR_DELAY'] > 0, 1, 0)
```

A binary column was created to identify delayed flights.

```python
# KPI Metrics
total_flights = df['FL_NUMBER'].count()
unique_airlines = df['AIRLINE'].nunique()
avg_delay = df['ARR_DELAY'].mean()

print(total_flights, unique_airlines, avg_delay)
```

These KPIs provide a quick overview of airline operations and delay patterns.

**Convert to datetime**

```python
# Convert to datetime
df['FL_DATE'] = pd.to_datetime(df['FL_DATE'])

# Extract useful features
df['YEAR'] = df['FL_DATE'].dt.year
df['MONTH'] = df['FL_DATE'].dt.month
df['DAY'] = df['FL_DATE'].dt.day
df['DAY_OF_WEEK'] = df['FL_DATE'].dt.day_name()
```
