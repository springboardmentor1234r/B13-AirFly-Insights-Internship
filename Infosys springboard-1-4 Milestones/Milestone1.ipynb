```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype('category')

for col in df.select_dtypes(include=['int64']).columns:
    df[col] = pd.to_numeric(df[col], downcast='integer')
```

```python
df.info()
```

```python
df.to_csv('../data/processed/flights_sample_3m_week1_cleaned.csv', index=False)
```

```python
df.columns
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['CANCELLATION_CODE'] = df['CANCELLATION_CODE'].cat.add_categories('Not Cancelled')
df['CANCELLATION_CODE'] = df['CANCELLATION_CODE'].fillna('Not Cancelled')
```

```python
df['CANCELLATION_CODE'].isnull().sum()
```

```python
delay_cols = [
    'DELAY_DUE_CARRIER',
    'DELAY_DUE_WEATHER',
    'DELAY_DUE_NAS',
    'DELAY_DUE_SECURITY',
    'DELAY_DUE_LATE_AIRCRAFT'
]

for col in delay_cols:
    df[col] = df[col].fillna(0)
```

```python
df.loc[df['CANCELLED'] == 1, 'DEP_DELAY'] = 0
df.loc[df['CANCELLED'] == 1, 'ARR_DELAY'] = 0
df['DEP_DELAY'] = df['DEP_DELAY'].fillna(0)
df['ARR_DELAY'] = df['ARR_DELAY'].fillna(0)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['CRS_ELAPSED_TIME'] = df['CRS_ELAPSED_TIME'].fillna(df['CRS_ELAPSED_TIME'].median())
```

```python
df['FL_DATE'] = pd.to_datetime(df['FL_DATE'])
```

```python
df['Year'] = df['FL_DATE'].dt.year
df['Month'] = df['FL_DATE'].dt.month
df['DayOfWeek'] = df['FL_DATE'].dt.day_name()
```

```python
df['DepHour'] = pd.to_datetime(df['DEP_TIME'], format='%H%M', errors='coerce').dt.hour
```

```python
df['Route'] = df['ORIGIN'].astype(str) + "-" + df['DEST'].astype(str)
```

```python
df['Route'] = df[['ORIGIN', 'DEST']].astype(str).agg('-'.join, axis=1)
```

```python
df[['ORIGIN','DEST','Route']].head()
```

```python
df = df[df['CANCELLED'] == 0]
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=[
    'DEP_TIME',
    'ARR_TIME',
    'AIR_TIME',
    'DEP_DELAY',
    'ARR_DELAY'
])
```

```python
df['FL_DATE'] = pd.to_datetime(df['FL_DATE'])
```

```python
df['DELAY_STATUS'] = df['ARR_DELAY'].apply(
    lambda x: 1 if x > 15 else 0
)
```

```python
df['TOTAL_DELAY'] = (
    df['DELAY_DUE_CARRIER'] +
    df['DELAY_DUE_WEATHER'] +
    df['DELAY_DUE_NAS'] +
    df['DELAY_DUE_SECURITY'] +
    df['DELAY_DUE_LATE_AIRCRAFT']
)
```

```python
df['DISTANCE_CATEGORY'] = pd.cut(
    df['DISTANCE'],
    bins=[0, 500, 1500, 3000, 10000],
    labels=['Short', 'Medium', 'Long', 'Very Long']
)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['AIRLINE_ENC'] = le.fit_transform(df['AIRLINE'])
df['ORIGIN_ENC'] = le.fit_transform(df['ORIGIN'])
df['DEST_ENC'] = le.fit_transform(df['DEST'])
```
