Data Cleaning

```python
# Drop Unnecessary Index Column
df = df.drop(columns=['index'])
```

**Convert Categorical Columns**

```python
categorical_cols = [
    'airline', 'flight', 'source_city',
    'departure_time', 'stops',
    'arrival_time', 'destination_city', 'class'
]

for col in categorical_cols:
    df[col] = df[col].astype('category')
```

**Convert Duration to Numeric**

```python
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')
```

Remove Duplicates

```python
df = df.drop_duplicates()
```

4.Feature Engineering
Create Route Column

```python
df['route'] = df['source_city'].astype(str) + " → " + df['destination_city'].astype(str)
```

**Convert days_left into Booking Urgency Category**

```python
df['booking_type'] = pd.cut(
    df['days_left'],
    bins=[0, 7, 30, 60, 365],
    labels=['Last_Minute', 'Short_Term', 'Mid_Term', 'Long_Term']
)
```

```python
#Save Cleaned Data
df.to_csv("airfly_cleaned_data.csv", index=False)
```
