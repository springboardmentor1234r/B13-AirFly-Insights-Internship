Top 10 route analysis

```python
top_routes = df['route'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_routes.plot(kind='bar')
plt.title("Top 10 Most Frequent Routes")
plt.xlabel("Route")
plt.ylabel("Number of Flights")
plt.xticks(rotation=45)
plt.show()
```

Average price by route

```python
route_price = df.groupby('route')['price'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
route_price.plot(kind='bar')
plt.title("Top 10 Expensive Routes")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```

price vs days left

```python
plt.figure(figsize=(8,5))
sns.lineplot(x='days_left', y='price', data=df)
plt.title("Price vs Days Left")
plt.xlabel("Days Left")
plt.ylabel("Price")
plt.show()
```

heat map:sorce vs destination

```python
route_matrix = pd.crosstab(df['source_city'], df['destination_city'])

plt.figure(figsize=(10,6))
sns.heatmap(route_matrix, cmap='coolwarm')
plt.title("Flight Density Heatmap")
plt.show()
```

duration vs price

```python
plt.figure(figsize=(8,5))
sns.regplot(x='duration', y='price', data=df, scatter_kws={'alpha':0.3})
plt.title("Duration vs Price (Trend)")
plt.show()
```

class wise price comparison
