Route & Airport-Level Analysis

```python
top_routes = df['Route'].value_counts().head(10)

plt.figure()
top_routes.plot(kind='bar')
plt.title("Top 10 Busiest Routes")
plt.xlabel("Route")
plt.ylabel("Number of Flights")
plt.show()
```

Delay by Route (Top Routes Only)

```python
top_routes_list = df['Route'].value_counts().head(10).index

route_delay = df[df['Route'].isin(top_routes_list)].groupby('Route')['ARR_DELAY'].mean()

plt.figure()
route_delay.plot(kind='bar')
plt.title("Average Delay for Top Routes")
plt.ylabel("Delay (minutes)")
plt.show()
```

Airport-Level Delay Analysis
Top Delay-Prone Origin Airports

```python
origin_delay = df.groupby('ORIGIN')['ARR_DELAY'].mean().sort_values(ascending=False).head(10)

plt.figure()
origin_delay.plot(kind='bar')
plt.title("Top Delay-Prone Origin Airports")
plt.ylabel("Average Delay")
plt.show()
```

Heatmap - Airport vs Month

```python
pivot = df.pivot_table(values='ARR_DELAY', index='ORIGIN', columns='Month')

plt.figure(figsize=(12,8))
sns.heatmap(pivot, cmap="coolwarm")

plt.title("Delay Heatmap (Airport vs Month)")
plt.show()
```

Seasonal & Cancellation Analysis
**Monthly Cancellation Trends**

```python
monthly_cancel = df.groupby('Month')['CANCELLED'].sum()

plt.figure()
monthly_cancel.plot(kind='line', marker='o')
plt.title("Monthly Cancellation Trends")
plt.ylabel("Number of Cancelled Flights")
plt.show()
```

# Cancellation Types Analysis

```python
cancel_cols = ['DELAY_DUE_CARRIER','DELAY_DUE_WEATHER','DELAY_DUE_NAS','DELAY_DUE_SECURITY']

cancel_mean = df[cancel_cols].mean()

plt.figure()
cancel_mean.plot(kind='bar')
plt.title("Cancellation/Delay Causes")
plt.ylabel("Average Delay")
plt.show()
```
