**AirFly Insights: Data Visualization and Analysis of Airline Operations**

***Milestone 1: Data Foundation and Cleaning***

Goal: Analyze airline delays / sales / traffic patterns


```python
#Import Libraries

import pandas as pd
import numpy as np
```

```python
#Load Dataset

df = pd.read_csv('data/Flight_Dataset.csv')

print("Initial Shape:", df.shape)
df.head()
```

```python
#Data Understanding

pd.set_option('display.max_columns', 40)

print("\nColumns:\n", df.columns)
print("\nData Info:")
df.info()

print("\nSummary Statistics:")
df.describe()
```

```python
#Duplicate Check

print("\nDuplicate Rows:", df.duplicated().sum())
```

```python
#Missing Values Analysis

print("\nMissing Values:\n")
print(df.isnull().sum())
```

```python
#Understanding Nulls (Important Insight)

print("\nSample rows where DEP_TIME is NULL:")
print(df[df['DEP_TIME'].isnull()].head())
```

```python
#Handling Cancellation Code

# Flights not cancelled → NaN in cancellation_code
df['CANCELLATION_CODE'] = df['CANCELLATION_CODE'].fillna('NC')
```

```python
#Removing Invalid CRS_ELAPSED_TIME

df.dropna(subset=['CRS_ELAPSED_TIME'], inplace=True)
```

```python
#Handling Delay Columns (Major Logic)

delays = [
    'DELAY_DUE_CARRIER', 'DELAY_DUE_WEATHER',
    'DELAY_DUE_NAS', 'DELAY_DUE_SECURITY',
    'DELAY_DUE_LATE_AIRCRAFT'
]
```

```python
#Flights arrived early/on-time → delay = 0
mask = (
    (df['CANCELLATION_CODE'] == 'NC') &
    (df['DIVERTED'] == 0) &
    (df['ARR_DELAY'] <= 0)
)

df.loc[mask, delays] = df.loc[mask, delays].fillna(0.0)
```

```python
#Investigating Operational Nulls

print("\nDEP_TIME NULL Analysis:")
print(df[df['DEP_TIME'].isna()][['CANCELLED','DIVERTED']].value_counts())

print("\nTAXI_IN NULL Analysis:")
print(df[df['TAXI_IN'].isna()][['CANCELLED','DIVERTED']].value_counts())

```

```python
#Removing Logical Errors

invalid_rows = df[
    (df['TAXI_IN'].isna()) &
    (df['CANCELLED'] == 0) &
    (df['DIVERTED'] == 0)
]

print("\nInvalid Rows Removed:", invalid_rows.shape)

df = df.drop(invalid_rows.index)
```

```python
#Handling Minor Delays (<=15 mins)

mask1 = (
    (df['CANCELLED'] == 0) &
    (df['DIVERTED'] == 0) &
    (df['ARR_DELAY'] <= 15)
)

df.loc[mask1, delays] = df.loc[mask1, delays].fillna(0.0)
```

```python
#Final Missing Values Check

print("\nRemaining Null Values:\n")
print(df.isnull().sum())
```

```python
#Extra Validation (ADDED)

print("\nFlights with Negative Arrival Delay (Early Arrival):")
print(df[df['ARR_DELAY'] < 0].shape)

print("\nFlights with Extreme Delay (>1000 mins):")
print(df[df['ARR_DELAY'] > 1000].shape)
```

```python
#Data Type Conversion

df['FL_DATE'] = pd.to_datetime(df['FL_DATE'])

df['CANCELLED'] = df['CANCELLED'].astype('int8')
df['DIVERTED'] = df['DIVERTED'].astype('int8')

cat_cols = [
    'AIRLINE','AIRLINE_DOT','AIRLINE_CODE',
    'ORIGIN','ORIGIN_CITY','DEST','DEST_CITY',
    'CANCELLATION_CODE'
]

for col in cat_cols:
    df[col] = df[col].astype('category')


# Convert float columns
minu_cols = [
    'DEP_DELAY','ARR_DELAY','TAXI_IN','TAXI_OUT',
    'CRS_ELAPSED_TIME','ELAPSED_TIME','AIR_TIME',
    'DELAY_DUE_CARRIER','DELAY_DUE_WEATHER',
    'DELAY_DUE_NAS','DELAY_DUE_SECURITY',
    'DELAY_DUE_LATE_AIRCRAFT'
]

for col in minu_cols:
    df[col] = df[col].astype('float32')


# Convert time columns
time_cols = [
    'CRS_DEP_TIME','DEP_TIME','CRS_ARR_TIME',
    'ARR_TIME','WHEELS_OFF','WHEELS_ON'
]

for col in time_cols:
    df[col] = df[col].astype('Int32')


# Optimize numeric columns
df['DISTANCE'] = df['DISTANCE'].astype('float32')
df['DOT_CODE'] = df['DOT_CODE'].astype('int32')
df['FL_NUMBER'] = df['FL_NUMBER'].astype('int32')

```

```python
#Feature Engineering

df['YEAR'] = df['FL_DATE'].dt.year.astype('int32')
df['MONTH'] = df['FL_DATE'].dt.month.astype('int8')
df['MONTH_NAME'] = df['FL_DATE'].dt.month_name().astype('category')
df['WEEKDAY'] = df['FL_DATE'].dt.day_name().astype('category')

# Departure Hour
df['DEP_HOUR'] = df['DEP_TIME'] // 100
df['DEP_HOUR'] = df['DEP_HOUR'].fillna(df['CRS_DEP_TIME'] // 100)
df['DEP_HOUR'] = df['DEP_HOUR'].astype('Int8')

# Route Column
df['ROUTE'] = (df['ORIGIN'].astype(str) + '-' + df['DEST'].astype(str)).astype('category')
```

```python
#Additional Features

# Delay Category
df['DELAY_STATUS'] = pd.cut(
    df['ARR_DELAY'],
    bins=[-1000, 0, 15, 60, 180, 10000],
    labels=['Early', 'On-Time', 'Minor Delay', 'Moderate Delay', 'Severe Delay']
)

# Weekend Indicator
df['IS_WEEKEND'] = df['WEEKDAY'].isin(['Saturday','Sunday']).astype('int8')
```

```python
#Final Dataset Info

print("\nFinal Shape:", df.shape)
df.info()
```

```python
#KPI Calculations

total_flights = len(df)
unique_airlines = df['AIRLINE'].nunique()

cancelled_pct = (df['CANCELLED'].sum() / total_flights) * 100
diverted_pct = (df['DIVERTED'].sum() / total_flights) * 100

late_departure_pct = (
    (df['DEP_DELAY'] > 0).sum() /
    df['DEP_DELAY'].notna().sum()
) * 100

avg_dep_delay = df[(df['DEP_DELAY'] > 0) & (df['CANCELLED'] == 0)]['DEP_DELAY'].mean()
avg_arr_delay = df[(df['ARR_DELAY'] > 0) & (df['CANCELLED'] == 0)]['ARR_DELAY'].mean()
```

```python
#KPI Output

print("\n========== KPI SUMMARY ==========")
print("Total Flights:", total_flights)
print("Unique Airlines:", unique_airlines)
print(f"Cancelled Flights: {cancelled_pct:.2f}%")
print(f"Diverted Flights: {diverted_pct:.2f}%")
print(f"Late Departures: {late_departure_pct:.2f}%")
print(f"Avg Departure Delay: {avg_dep_delay:.2f} mins")
print(f"Avg Arrival Delay: {avg_arr_delay:.2f} mins")
```

```python
#Save Cleaned Dataset

df.to_csv('cleaned_flights_sample_final.csv', index=False)

print("\n✅ Dataset saved successfully!")
```
