Correlation Analysis

```python
corr = df[['duration', 'days_left', 'price']].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

Top airlines by revenue

```python
revenue = df.groupby('airline')['price'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
revenue.plot(kind='bar')
plt.title("Total Revenue by Airline (Approx)")
plt.ylabel("Total Price")
plt.show()
```

Price distribution by departure time

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='departure_time', y='price', data=df)
plt.title("Price by Departure Time")
plt.xticks(rotation=45)
plt.show()
```

stops vs duration

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='stops', y='duration', data=df)
plt.title("Stops vs Duration")
plt.show()
```
